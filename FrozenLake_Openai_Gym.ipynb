 #### Markov decision process (MDP) consists of the following:
- States: A set of states present in the environment.
- Actions: A set of actions that the agent can perform in each state.
- Transition probability: The transition probability is denoted by ùëÉ(s‚Ä≤|ùë†, a). It implies the probability of moving from a state s to the state ùë†‚Ä≤  while performing an action a.
- Reward function: The reward function is denoted by ùëÖ(s, a, s‚Ä≤) It implies 
the reward the agent obtains moving from a state s to the state ùë†‚Ä≤ while  performing an action a.

### Generating an episode

```python
import gym

env = gym.make("FrozenLake-v0")

env.reset()

print('Time Step 0 :')
env.render()

# number of time steps
num_timesteps = 20

for t in range(num_timesteps):
  random_action = env.action_space.sample()
  (next_state, reward, done, info) = env.step(random_action)
  print ('Time Step {} :'.format(t+1))
  env.render()
  if done==True:
        break
```

Instead of generating one episode, we can also generate a series of episodes by taking some random action in each state:

```python
import gym
env = gym.make("FrozenLake-v0")

num_episodes = 5
num_timesteps = 20 
for i in range(num_episodes):
    
    state = env.reset()
    print('Episode number {} :'.format(i+1))
    env.render()
    for t in range(num_timesteps):
        random_action = env.action_space.sample()
        
        new_state, reward, done, info = env.step(random_action)
        print ('Time Step {} :'.format(t+1))

        env.render()
        if done:
            break
```
